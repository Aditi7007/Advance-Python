```python
import pandas as pd
import numpy as np

technologies = {
    'Courses': ["Spark", "PySpark", "Hadoop", "Python", "Pandas", None, "Spark", "Python"],
    'Fee': [22000, 25000, 23000, 24000, np.nan, 25000, 25000, 22000],
    'Duration': ['30day', '50days', '55days', '40days', '60days', '35day', '', '50days'],
    'Discount': [1000, 2300, 1000, 1200, 2500, 1300, 1400, 1600]
}

row_labels = ['r0', 'r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7']
df = pd.DataFrame(technologies, index=row_labels)

# Display the DataFrame for reference
print(df)
```

1.Describe dataframe for all numeric column

```python
numeric_description = df.describe()
print(numeric_description)
```

2.. Execute a query that contains records of only spark course.

```python
spark_courses = df[df['Courses'] == 'Spark']
print(spark_courses)
```

3. Display the records whose Course Fee between 23000 and 24000.

```python
fee_range_courses = df[(df['Fee'] >= 23000) & (df['Fee'] <= 24000)]
print(fee_range_courses)
```

4. Display all record other than Spark course

```python
non_spark_courses = df[df['Courses'] != 'Spark']
print(non_spark_courses)
```

5. Display records whose Discount between 1000 and 2000.

```python
discount_range_courses = df[(df['Discount'] >= 1000) & (df['Discount'] <= 2000)]
print(discount_range_courses)
```
