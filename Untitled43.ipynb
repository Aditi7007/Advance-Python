```python
#Question15
import pandas as pd

# Sample dataset for illustration purposes
data = {
    'date': ['2025-04-01', '2025-04-02', '2025-04-03', '2025-04-04'],
    'sales': [1200, 850, 1500, 950],
    'region': ['North', 'South', 'North', 'East']
}

# Create a DataFrame
df = pd.DataFrame(data)

# 1. Convert the date column to a datetime format
df['date'] = pd.to_datetime(df['date'])

# Print the output after converting date
print("Converted Date Column:")
print(df)

# 2. Calculate the total sales per region
total_sales_per_region = df.groupby('region')['sales'].sum().reset_index()

# Print the total sales per region
print("\nTotal Sales per Region:")
print(total_sales_per_region)

# 3. Filter the rows to show only the sales that are greater than 1000
filtered_sales = df[df['sales'] > 1000]

# Print the output after filtering sales > 1000
print("\nSales Greater Than 1000:")
print(filtered_sales)
```
